```python
import matplotlib.pyplot as plt
import numpy as np

def find(txt, tar):
    idx = txt.find(tar)
    rtxt = txt[idx + len(tar):]
    return float(rtxt.split(', ')[0])
```

```python
losses = []
rvalues = []
rrmses = []
tlosses = []
with open('../deep_emb_01.log', 'r') as f:
    lines = f.readlines()
    print(len(lines))
    for line in lines:
        if 'validation' in line:
            tar = 'loss = '
            loss = find(line, tar)
            tar = 'rvalue = '
            rvalue = find(line, tar)
            tar = 'rrmse = '
            rrmse = find(line, tar)
            
            losses.append(loss)
            rvalues.append(rvalue)
            rrmses.append(rrmse)
        elif 'epoch' in line and 'lr' in line:
            tar = 'loss = '
            loss = find(line, tar)
            tlosses.append(loss)
            

losses = np.array(losses)
rvalues = np.array(rvalues)
rrmses = np.array(rrmses)
tlosses = np.array(tlosses)

# print(losses.argmin(), losses.min())
# print(rrmses.argmin(), rrmses.min())
# print(rvalues.argmax(), rvalues.max())
idx = rvalues.argmax()
print('at epoch {}, loss = {}, rvalue = {}, rrmse = {}'.format(
    idx, losses[idx], rvalues[idx], rrmses[idx]
))

plt.plot(tlosses)
plt.show()

plt.plot(losses)
plt.show()
```
