```python
import pandas as pd
import numpy as np

import os
```

```python
base_dir = "../dataset/_datasets/cluster_res"
base_file = "seq_dg_v02_1.txt"
fpath = os.path.join(base_dir, base_file)
base_df = pd.read_csv(fpath, sep='\t')
base_df.head(3)

uniq_key_complex = np.unique(base_df['key_complex'].values)
mask = base_df['exp_id'].isin(uniq_key_complex)
base_df = base_df[mask]
print(len(base_df))
```

```python
list_data_size = []
cluster_cnt = 0
for key, grouped_df in base_df.groupby('base_class'):
    cluster_cnt += 1
    if len(grouped_df) > 10:
        list_data_size.append( len(grouped_df) )
cluster_cnt, len(list_data_size), list_data_size 
```

```python
max_len = 0
clu_df_by_key = dict()
largest_cluster_df = None
for key, grouped_df in base_df.groupby('base_class'):
#     if len(grouped_df) > max_len:
#         max_len = len(grouped_df)
#         largest_cluster_df = grouped_df
    clu_df_by_key[key] = grouped_df
    uniq_proteins = np.unique(grouped_df['protein_sequence'].values)
    if len(uniq_proteins) > max_len:
        max_len = len(uniq_proteins)
        largest_cluster_df = grouped_df
print(len(largest_cluster_df))
largest_cluster_df
```

```python
uniq_proteins = np.unique(largest_cluster_df['protein_sequence'].values)
uniq_nu_acids = np.unique(largest_cluster_df['nucleotide_sequence'].values)

uniq_uniprot_ids = np.unique(largest_cluster_df['UniProt'].values)
# len(uniq_proteins), len(uniq_nu_acids), len(uniq_uniprot_ids), len(largest_cluster_df)
# (47, 3, 1, 51)

seq2seq_id = {seq: i for i, seq in enumerate(uniq_proteins)}
na2na_id = {seq: i for i, seq in enumerate(uniq_nu_acids)}
seq_ids = [seq2seq_id[seq] for seq in largest_cluster_df['protein_sequence'].values]
na_ids = [na2na_id[seq] for seq in largest_cluster_df['nucleotide_sequence'].values]
print(seq_ids)
print(na_ids)
```

```python
prot_series = largest_cluster_df['protein_sequence'].apply(lambda k : seq2seq_id[k])
na_series = largest_cluster_df['nucleotide_sequence'].apply(lambda k : na2na_id[k])
prot_series = prot_series.astype('int32')
na_series = na_series.astype('int32')
out_df = pd.concat([na_series, prot_series, largest_cluster_df['dG']], axis=1)

out_df = out_df.reset_index()

out_df = out_df.sort_values(by=['nucleotide_sequence', 'protein_sequence'])
# out_df
# df.astype({'col1': 'int32'}).dtypes
out_df
```
